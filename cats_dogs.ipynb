```python
import numpy as np
import os
import matplotlib.pyplot as plt
import pandas as pd
import random 
import pickle
import cv2
```

```python
DIRECTORY = r"C:\Users\RADHIKA RAJESH VERMA\Desktop\PR_Py_.py\GIT\practice\classification_cats_dogs\train"
CATEGORIES = ['cats','dogs']
```

```python
data = []
for category in CATEGORIES:
    folder = os.path.join(DIRECTORY, category)
    label = CATEGORIES.index(category)
    # print (folder)
    for img in os.listdir(folder):
        img_path = os.path.join(folder, img)
        # print(img_path)
        # break
        img_arr = cv2.imread(img_path)
        img_arr = cv2.resize(img_arr, (1000,1000))
        plt.imshow(img_arr)
        break        
        data.append([img_arr, label])
```

```python
len(data)
```

```python
random.shuffle(data)
```

```python
x = []
y = []

for features, labels in data:
    x.append(features)
    y.append(labels)
    
```

```python
x = np.array(x)
y = np.array(y)
```
